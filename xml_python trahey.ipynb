```python
with urllib.request.urlopen(url_to_load) as f:
    tei_as_string = f.read().decode ('utf-8')
    
start_index = tei_as_string.find ('<term type="shape"')

tmp_end_index = tei_as_string [start_index:].find('>')

end_index = start_index + tmp_end_index + 1 

print("The first term type=shape element begins at index" + str(start_index))
print("The first term type=shape element ends at index" + str(end_index))
```
